```python
# creating a subset dataframe for feautures of interest. 
cols=['MemberKey', 'EmploymentStatus', 'LoanStatus', 'Occupation', 'BorrowerState', 'BorrowerAPR', 'BorrowerRate', 'StatedMonthlyIncome', 'LoanOriginalAmount', 'Term', 'IncomeRange', 'ListingCategory (numeric)',  'CreditScoreRangeLower', 'CreditScoreRangeUpper', 'ProsperRating (numeric)', 'TotalProsperLoans', 'LoanOriginationDate', 'LoanOriginationYear', 'MonthlyLoanPayment']
df_sub = loan_df[cols]
df_sub.head()
```

#### How is employment status distributed?

```python
plt.figure(figsize = [12, 8])
color_palt = sb.color_palette()[6]
type_order = df_sub['EmploymentStatus'].value_counts().index
sb.countplot(data= df_sub, y= 'EmploymentStatus', color = color_palt, order = type_order)
plt.title("Distribution of Employment Status")
plt.xlabel("Frequency")
plt.ylabel("Emloyment status");
```

### Insight:
> * Most borrowers are employed. This makes sense as it would be hard or impossible to get a loan without a job.

### Which state is the top loan borrowing state?

```python
# Plotting the distribution for the listed borrowing states
plt.figure(figsize = [12, 8])
color_paltt = sb.color_palette()[6]
type_order = df_sub['BorrowerState'].value_counts().index
sb.countplot(data = df_sub, x = 'BorrowerState', color = color_paltt, order = type_order)
plt.ylabel('Frequency')
plt.title('Borrowers States')
plt.xticks(rotation = 90)
;
```

### Insight:

> * Most borrowers are from **CA (CALIFORNIA)** state. This may be attributed to the fact that Prosper company is in California.

#### In which occupation is there likely to be more borrowers?

```python
#plotting the distribution of the boerrower's occupations
plt.figure(figsize = [12, 8])
color_paltt = sb.color_palette()[6]
order = df_sub['Occupation'].value_counts().iloc[:15].index
sb.countplot(data= df_sub, y= 'Occupation', color = color_paltt, order = order)
plt.title("Distribution of the top 15 Borrowers Occupation")
plt.xlabel("Frequency")
plt.ylabel("Occupation");
```

### Insight:

#### How is the distribution rate of the Borrower's rate?

```python
#plotting histogran to view borrower's rate distribution.

plt.figure(figsize = [12, 8])
color_paltt = sb.color_palette()[6]
bin_size = np.arange(df_sub.BorrowerRate.min(), df_sub.BorrowerRate.max()+0.01,0.01)
plt.hist(data = df_sub, x = 'BorrowerRate', bins = bin_size, color = color_paltt)
plt.xlabel('Borrower Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Borrower Rate');
```

## Bivariate Exploration

> In this section, I investigate relationships between pairs of variables in the 
data.

#### What is the realtionship between the numeric variables, LoanOriginalAmount, BorrowerAPR, BorrowerRate and Term ?

```python
#group numeric variables into one 

num_var = ['LoanOriginalAmount', 'BorrowerAPR', 'BorrowerRate', 'Term']
# use a heat map to dispaly and visualize correlation between numeric variables

plt.figure(figsize= (10, 8))
sb.heatmap(data=df_sub[num_var].corr(), annot=True, fmt='.2f', 
            cmap='vlag_r', vmin=-1, vmax=1, linewidth=3)
plt.xticks(rotation=90)
plt.yticks(rotation = 45);
```

### Observation:
> * There is a strong correlation between between the Borrowers rate and Borrower APR
* The moderate correlation observed  between Term and the Loan original amount could be attributed to the fact that a borrower takes more time time to pay for higher loan.

#### What is the relationship between loan original amount and borrower's annual percentage rate?

```python
plt.figure(figsize = [12, 8])
color_paltt = sb.color_palette()[6]
sb.regplot(data = df_sub, x = 'LoanOriginalAmount', y = 'BorrowerAPR', scatter_kws={'alpha':0.01},color = color_paltt)
plt.title('Loan Original Amount vs BorrowerAPR');
plt.xlabel('Loan amount($)')
```

### Observation:
> * There exists a negative relationship between Loan original amount and borrowers APR.
* The smaller the amount of loan requested by the borrower, the higher the interest rate.

#### What is the influence of the Employment status and Term on BorrowerAPR, Loan original amount ?

```python
#box plots to visualize numeric and categorical variable relationship

x_cols = ['EmploymentStatus']
y_cols = ['LoanOriginalAmount', 'BorrowerAPR']


#def boxgrid function
def boxgrid(x, y, **kwargs):
    color_pltt = sb.color_palette()[6]
    sb.boxplot(x, y, color = color_pltt)
    plt.xticks(rotation=90)


g = sb.PairGrid(data = df_sub, y_vars = y_cols, x_vars = x_cols, aspect = 3.5,)
g.map(boxgrid)
plt.subplots_adjust(hspace=1.0, wspace=0.6)
plt.xticks(rotation=90);

```

### Observation:
> * Borrowers who have jobs and are fulltime employed get higher loan amounts with lower Annual Percentage Rate.
